```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('mushrooms.csv')
df.head()
```

```python
df.info()
```

```python
for col in df.columns:
    print(df[col].value_counts())
```

All are categorical features and are nominal

### Checking for Class Imbalance

```python
df['class'].value_counts()
```

Class is balances since there is not a huge difference in the counts in each class

### Renaming and replacing Target variable

```python
mapping = {"e": 1, "p": 0}

df.rename({'class':'edible'}, axis=1, inplace=True)
df['edible'] = df['edible'].replace(mapping)
```

```python
df.head()
```

### Create X and y variables

```python
X = df.drop('edible', axis=1)
y = df['edible']
```

### Convert object to integers

```python
# OneHotEncoding to X
X = pd.get_dummies(X, drop_first=True )
X.head()
```

```python
df.head()
```

```python
df
```

## Create training and testing data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state = 42)
```

```python
X_train.shape, X_test.shape
```

### Decision Tree Classifier with criterion entropy

```python
from sklearn.tree import DecisionTreeClassifier

clf_en = DecisionTreeClassifier( criterion='entropy', max_depth = 3, random_state = 0)

clf_en.fit(X_train, y_train)
```

#### Predict the Test set results with criterion entropy

```python
y_pred_en = clf_en.predict(X_test)
```

#### Check accuracy score with criterion entropy

```python
from sklearn.metrics import accuracy_score

print('Model accuracy score with criterion entropy: {0:0.4f}'. format(accuracy_score(y_test, y_pred_en)))
```

#### Compare the train-set and test-set accuracy

```python
y_pred_train_en = clf_en.predict(X_train)

print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train_en)))
```

The training-set accuracy score is 0.9638 while the test-set accuracy to be  0.9571. 
<br> These two values are quite comparable. So, there is no sign of overfitting.

## Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred_en)

print('Confusion matrix\n\n', cm)
```

## Classification Report

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_en))
```

# Random Forest Classifier

### Random Forest Classifier model with default parameters

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(random_state=0)
```

```python
rfc.fit(X_train, y_train)

y_pred = rfc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print('Model accuracy score with 10 decision-trees : {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```

```python
y_pred_train = rfc.predict(X_train)

print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train)))
```

### Random Forest Classifier model with parameter n_estimators=100

```python
rfc_100 = RandomForestClassifier(n_estimators= 100, random_state=0)

rfc_100.fit(X_train, y_train)

y_pred_100 = rfc.predict(X_test)
```

```python
print('Model accuracy score with 100 decision-trees : {0:0.4f}'. format(accuracy_score(y_test, y_pred_100)))
```

```python
y_pred_train = rfc_100.predict(X_train)

print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train)))
```

###  Find important features with Random Forest model

```python
clf = RandomForestClassifier(n_estimators= 100, random_state=0)

clf.fit(X_train, y_train)
```

```python
# view feature scores

feature_scores = pd.Series(clf.feature_importances_,index=X_train.columns).sort_values(ascending=False)
feature_scores
```
